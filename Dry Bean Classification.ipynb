```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
df['Class']= le.fit_transform(df['Class'])
df
```

```python
x= df.drop('Class',axis=1)
y= df['Class']
```

```python
x= x.values
y=y.values
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size= 0.2, random_state= 100)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree= DecisionTreeClassifier(max_depth= 10)
dtree.fit(x_train,y_train)
y_train_predicted= dtree.predict(x_train)
y_test_predicted= dtree.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix,classification_report
```

```python
print('Model Performance')
print('Training Accuracy: ',accuracy_score(y_train,y_train_predicted))
print(confusion_matrix(y_train,y_train_predicted))
print(classification_report(y_train,y_train_predicted))
```

```python
print('Model Performance')
print('Testing Accuracy: ',accuracy_score(y_test,y_test_predicted))
print(confusion_matrix(y_test,y_test_predicted))
print(classification_report(y_test,y_test_predicted))
```

```python
import numpy as np
import gradio as gr

def prediction(Area, Extent, Solidity, ShapeFactor3, ShapeFactor4):
    input_data = np.array([Area, Extent, Solidity, ShapeFactor3, ShapeFactor4])
    input_data = input_data.reshape(1, -1)  # Ensure 2D shape
    
    prediction = rf.predict(input_data)  # Ensure rf is trained before calling predict()
    
    # Use inverse_transform() instead of prediction_transform()
    return le.inverse_transform(prediction)[0]

# Test function
print(prediction(3.53640000e+04, 7.92168810e-01, 9.88898521e-01, 9.111973933e-01, 9.98391922e-01))
```

```python
inputs = [
    gr.Number(label = "Enter Area"),
    gr.Number(label = "Enter Extent"),
    gr.Number(label = "Enter Solidity"),
    gr.Number(label = "Enter ShapeFactor3"),
    gr.Number(label = "Enter ShapeFactor4")
]
```

```python
output = gr.Textbox(label = 'Prediction')
```

```python
gr.Interface(fn = prediction, inputs = inputs, outputs = output,
    title = "Beans classification",
    description = "Enter the input data ",
    ).launch(share = True)
```

# Bagging 

```python
from sklearn.ensemble import VotingClassifier 
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier 
from sklearn.svm import SVC 
model1 = LogisticRegression()
model2 = DecisionTreeClassifier()
model3 = SVC()
```

```python
ensemble_classifier = VotingClassifier(
    estimators = [ ('lr', model1), ('dt',model2), ('svc',model3)], voting = 'hard'
)
```

```python
ensemble_classifier.fit(x_train, y_train)
y_train_predicted= ensemble_classifier.predict(x_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report

print("Model Performance")
print("Training accuracy: ", accuracy_score(y_test, y_test_predicted))
# print("Testing precsision: ", precision_score(y_test, y_test_predicted))
# print("Testing recall: ", recall_score(y_test, y_test_predicted))
print(confusion_matrix(y_test, y_test_predicted))
print(classification_report(y_test, y_test_predicted))
```
